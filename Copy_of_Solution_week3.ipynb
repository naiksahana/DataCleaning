# Age

```python
avg_age = data.age.mean()
```

```python
data['age'] = data['age'].fillna(avg_age)
data['age'] = data['age'].round(decimals=1)
data['age']
```

```python
data['age'].isna().sum()
```

# Salary

```python
data['salary'] = data['salary'].apply(lambda x:x.replace('$',''))
data['salary'] = data['salary'].apply(lambda x:x.replace('k','000'))
```

```python
data['age'] = data['age'].astype(int)
```

established

```python
data['established'].isna().sum()
```

```python
import pandas as pd

# Assuming 'data' is your pandas DataFrame

# Convert the 'established' column to numeric data type
data['established'] = pd.to_numeric(data['established'], errors='coerce')

# Fill NaN values with the mean and round to one decimal place
data['established'] = data['established'].fillna(data['established'].mean()).round(decimals=1)

# Print the converted and processed 'established' column
print(data['established'])
```

Location column:

cleaning the location and splitting the city from the city short name

```python
#cleaning the location and splitting the city from the city short name
data['location']=data['location'].str.replace("Australia Aus","Austalia.Aus")
data['location_city']=data['location'].str.split(",",expand=True).get(0)
data['city_sign']=data['location'].str.split(",",expand=True).get(1)
data.drop('location',inplace=True,axis=1)
```
