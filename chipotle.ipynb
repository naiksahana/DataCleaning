Extract individual item from choice description:

```python
def extract_items(choice_description):
  if pd.isnull(choice_description):
    return[]
  else:
      return[item.strip("[]")for item in choice_description.split(",")]
```

apply the extract_items fun to create a new column 'parsed items':

```python
data['parsed_items']=data['choice_description'].apply(extract_items)
```

creating new column for each item:

```python
unique_items=set(item for sublist in data['parsed_items'] for item in sublist)
for item in unique_items:
  data[item]=data['parsed_items'].apply(lambda x:1 if item in x else 0)
```
